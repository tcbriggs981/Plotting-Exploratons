```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
population = pd.read_csv('datasets/PEP_2017_PEPANNRES_with_ann.csv')

population_updates = population.drop(['Id', 'Id2', 'April 1, 2010 - Census', 'April 1, 2010 - Estimates Base'], axis=1)

population_indexed = population_updates.set_index('Geography')

population_indexed.head()
```

```python
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]

plt.figure(figsize=(20, 5))

plt.subplot(1,4,1)
plt.plot(x, population_indexed.loc['United States'])
plt.title('US Population (millions)')
plt.subplot(1,4,2)
plt.plot(x, population_indexed.loc['New York'])
plt.title('New York Population (millions)')
plt.subplot(1,4,3)
plt.plot(x, population_indexed.loc['Ohio'])
plt.title('Ohio Population (millions)')
plt.subplot(1,4,4)
plt.plot(x, population_indexed.loc['Texas'])
plt.title('Texas Population (millions)')
plt.show()
```

```python
only_states = population_indexed.drop('United States')

plt.figure(figsize=(50, 10))
plt.plot(only_states['Population Estimate (as of July 1) - 2017'], alpha=.75)
plt.plot(only_states['Population Estimate (as of July 1) - 2010'], alpha=.75)
plt.title('US Population Change by State: 2010–2017')
plt.legend(['2017', '2010'])
plt.show()
```

```python
life_ex = pd.read_csv('datasets/NCHS_-_Death_rates_and_life_expectancy_at_birth.csv')

life_ex.head()
```

```python
black_all = life_ex.loc[(life_ex['Race']=='Black') & (life_ex['Sex']=='Both Sexes')]
white_all = life_ex.loc[(life_ex['Race']=='White') & (life_ex['Sex']=='Both Sexes')]

male_all = life_ex.loc[(life_ex['Race']=='All Races') & (life_ex['Sex']=='Male')]
female_all = life_ex.loc[(life_ex['Race']=='All Races') & (life_ex['Sex']=='Female')]

white_male = life_ex.loc[(life_ex['Race']=='White') & (life_ex['Sex']=='Male')]
black_female = life_ex.loc[(life_ex['Race']=='Black') & (life_ex['Sex']=='Female')]

plt.figure(figsize=(10, 20))

plt.subplot(3,1,1)
plt.plot(white_all['Year'], white_all['Average Life Expectancy (Years)'])
plt.plot(black_all['Year'], black_all['Average Life Expectancy (Years)'])
plt.ylabel('Average Life Expectancy (Years)')
plt.xlabel('Year')
plt.title('US Life Expectancy Among White & Black Citizens')
plt.legend(['White', 'Black'])

plt.subplot(3,1,2)
plt.plot(male_all['Year'], male_all['Average Life Expectancy (Years)'])
plt.plot(female_all['Year'], female_all['Average Life Expectancy (Years)'])
plt.ylabel('Average Life Expectancy (Years)')
plt.xlabel('Year')
plt.title('US Life Expectancy - Male & Female')
plt.legend(['Male', 'Female'])

plt.subplot(3,1,3)
plt.plot(white_male['Year'], white_male['Average Life Expectancy (Years)'])
plt.plot(black_female['Year'], black_female['Average Life Expectancy (Years)'])
plt.ylabel('Average Life Expectancy (Years)')
plt.xlabel('Year')
plt.title('US Life Expectancy - White Male vs Black Female')
plt.legend(['White Male', 'Black Female'])

plt.show()
```
